1.	Función que calcule la molaridad de una solución: Gramos, volumen mL a L, peso molecular


```python
def calcular_molaridad(gramos, volumen_mL, peso_molecular):

  volumen_1=volumen_mL/1000
  molaridad=gramos/(volumen_1*peso_molecular)
  return molaridad

#Datos del ejemplo
gramos_soluto=10
volumen_mL=210
peso_molecular=74.093

#calcular la molaridad
molaridad=calcular_molaridad(gramos_soluto, volumen_mL, peso_molecular)
print(molaridad)
```

```python
import math

def calcular_pH(concentracion_H):
  if concentracion_H <=0:
    raise ValueError("La concentración de H debe ser positiva")
  pH= -math.log10(concentracion_H)
  return pH

#Datos
concentracion_H=0.2
print(calcular_pH(concentracion_H))
```

```python
from builtins import print

def calcular_masa_molar(gramos, volumen, temperatura_C, presion_atm):
  temperatura_K=temperatura_C+273.15 #temperatura de °C a K
  R = 0.0821 #(L.atm/(mol.K))
  masa_molar=(gramos * R * temperatura_K) /(volumen*presion_atm)
  return masa_molar

#Datos
gramos=2.5
volumen=1.5
temperatura_C=25
presion_atm=1.0
print(calcular_masa_molar(gramos, volumen, temperatura_C, presion_atm))
```

```python
def calcular_concentraciones_buffer(pH_deseado, pKa, volumen_ml):

    # Convertir volumen de mL a L
    volumen_l = volumen_ml / 1000
    # Relación base/ácido
    relacion_base_acido = 10**(pH_deseado - pKa)
    # Calcular las concentraciones
    # Supongamos que la suma de las concentraciones de ácido y base es 1 mol/L (puedes ajustar según sea necesario)
    total_concentracion = 1  # mol/L
    concentracion_acido = total_concentracion / (1 + relacion_base_acido)
    concentracion_base = total_concentracion - concentracion_acido

    return {
        "Concentración ácido (mol/L)": concentracion_acido,
        "Concentración base conjugada (mol/L)": concentracion_base,
        "Volumen solución (L)": volumen_l
    }

# Datos de entrada
pH_deseado = 7.4  # pH deseado del buffer
pKa = 6.8  # pKa del ácido monoprótico
volumen_solucion = 1000  # Volumen total de la solución en mL

# Calcular concentraciones
resultado = calcular_concentraciones_buffer(pH_deseado, pKa, volumen_solucion)

# Mostrar el resultado
print(f"Datos para preparar el buffer:")
print(f" - pH deseado: {pH_deseado}")
print(f" - pKa del ácido: {pKa}")
print(f" - Volumen total: {resultado['Volumen solución (L)']} L")
print(f"\nConcentraciones necesarias:")
print(f" - Ácido: {resultado['Concentración ácido (mol/L)']:.4f} mol/L")
print(f" - Base conjugada: {resultado['Concentración base conjugada (mol/L)']:.4f} mol/L")
```

```python
def calcular_rendimiento_porcentual(masa_teorica, masa_experimental):
    if masa_teorica <= 0:
        raise ValueError("La masa teórica debe ser positiva")
    rendimiento=(masa_experimental/masa_teorica)*100
    return rendimiento

#Datos
masa_teorica=5.9
masa_experimental=5.1
print(calcular_rendimiento_porcentual(masa_teorica, masa_experimental))
```

```python
def calcular_calor_especifico(masa, delta_temp, calor):
   if masa <= 0 or delta_temp == 0:
        raise ValueError("La masa debe ser mayor que 0 y el cambio de temperatura no puede ser 0.")
   c = calor / (masa * delta_temp)
   return c

#Datos
masa_metal = 80.0  # Masa del metal en gramos
delta_temp_metal = 20.0  # Cambio de temperatura en °C
calor_metal = 900.8  # Calor absorbido en julios (J)
print(calcular_calor_especifico(masa_metal, delta_temp_metal, calor_metal))
```

```python
def calcular_presion_osmotica(concentracion_g_l, peso_molecular, temperatura_c, R=0.0821):
  # Convertir temperatura de °C a K
    temperatura_k = temperatura_c + 273.15
    # Calcular la molaridad
    molaridad = concentracion_g_l / peso_molecular
    # Calcular la presión osmótica
    presion_osmotica = molaridad * R * temperatura_k
    return presion_osmotica

#Datos
concentracion = 15.0  # Concentración en g/L
peso_molecular = 84.9947 # Peso molecular del soluto NaNO3 en g/mol
temperatura = 25.0  # Temperatura en °C
print(calcular_presion_osmotica(concentracion, peso_molecular, temperatura))
```

```python
def calcular_producto_solubilidad(solubilidad, estequiometria):
   if solubilidad <= 0:
        raise ValueError("La solubilidad debe ser mayor que 0.")

    # Calcular Kps según la estequiometría
   Kps = 1
   for coef in estequiometria:
        Kps *= (coef * solubilidad)**coef
   return Kps

#Datos
solubilidad = 0.02  # Solubilidad de la sal en mol/L
estequiometria = [1, 2]  # Estequiometría para AB2 (BaF2)
print(calcular_producto_solubilidad(solubilidad, estequiometria))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Seno de x')
plt.title('Gráfico de Seno', fontsize=16)
plt.xlabel('Valores de x', fontsize=12)
plt.ylabel('Seno(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
